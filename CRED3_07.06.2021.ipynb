### Gain Calculation

#### Variance vs. mean

```python
x = np.arange(0, cube_1.shape[0])
y = np.arange(0, cube_1.shape[1])
gain = np.zeros((len(x),len(y)))
for i in x:
    for j in y:
        var_arr = np.array([var_1[i,j], var_2[i,j], var_3[i,j], var_4[i,j], var_5[i,j], var_6[i,j], 
                            var_7[i,j], var_8[i,j], var_9[i,j], var_10[i,j], var_11[i,j], var_12[i,j], 
                            var_13[i,j], var_14[i,j], var_15[i,j], var_16[i,j], var_17[i,j], var_18[i,j], 
                            var_19[i,j], var_20[i,j], var_21[i,j]])
        mean_arr = np.array([mean_1[i,j], mean_2[i,j], mean_3[i,j], mean_4[i,j], mean_5[i,j], mean_6[i,j], 
                            mean_7[i,j], mean_8[i,j], mean_9[i,j], mean_10[i,j], mean_11[i,j], mean_12[i,j], 
                            mean_13[i,j], mean_14[i,j], mean_15[i,j], mean_16[i,j], mean_17[i,j], mean_18[i,j], 
                            mean_19[i,j], mean_20[i,j], mean_21[i,j]])
        res = linregress(mean_arr, var_arr)
        m = res.slope
        gain[i,j] = 1/m
```

```python
gain = np.median(gain, axis = None)
print('Gain = %s e-/ADU'%(gain))
```

## Gain Calculation

#### Variance vs. mean

```python
x = np.arange(0, cube_1.shape[0])
y = np.arange(0, cube_1.shape[1])
gain = np.zeros((len(x),len(y)))
for i in x:
    for j in y:
        var_arr = np.array([var_1[i,j], var_2[i,j], var_3[i,j], var_4[i,j], var_5[i,j], var_6[i,j], 
                            var_7[i,j], var_8[i,j], var_9[i,j], var_10[i,j], var_11[i,j], var_12[i,j], 
                            var_13[i,j], var_14[i,j], var_15[i,j], var_16[i,j], var_17[i,j], var_18[i,j], 
                            var_19[i,j]])
        mean_arr = np.array([mean_1[i,j], mean_2[i,j], mean_3[i,j], mean_4[i,j], mean_5[i,j], mean_6[i,j], 
                            mean_7[i,j], mean_8[i,j], mean_9[i,j], mean_10[i,j], mean_11[i,j], mean_12[i,j], 
                            mean_13[i,j], mean_14[i,j], mean_15[i,j], mean_16[i,j], mean_17[i,j], mean_18[i,j], 
                            mean_19[i,j]])
        res = linregress(mean_arr, var_arr)
        m = res.slope
        gain[i,j] = 1/m
```
